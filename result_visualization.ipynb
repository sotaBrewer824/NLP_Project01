```python
import matplotlib.pyplot as plt
import numpy as np

percentages = [70.6, 68.3, 83.1, 81.5, 73.8]

labels = ["Yolo-v10", "LayoutXLM-384", "LayoutXLM-512", "LiLT-384", "LiLT-512"]


colors = plt.cm.tab10(np.linspace(0, 1, len(percentages)))

# draw the bar
plt.bar(labels, percentages, color=colors)



plt.title('Accuracy of different models in De', fontsize=12)
plt.xlabel('Different Models', fontsize=10)
plt.ylabel('Percentages (%)', fontsize=10)


plt.ylim(0, 110)

plt.xticks(fontsize=8)
plt.yticks(fontsize=8)


plt.show()
```

```python
import matplotlib.pyplot as plt

percentages = [81.5, 73.8]

labels = ["384", "512"]


colors = ['green' if p >= 80 else 'red' for p in percentages]

# draw the bar
plt.bar(labels, percentages, color=colors)

plt.title('Accuracy of different token size of LiLT in De', fontsize=12)
plt.xlabel('Different token size', fontsize=10)
plt.ylabel('Percentages (%)', fontsize=10)


plt.ylim(0, 110)

plt.xticks(fontsize=8)
plt.yticks(fontsize=8)


plt.show()
```
